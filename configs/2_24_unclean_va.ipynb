```python
# small dataset is faster and allows easier comparison of clean and raw since we have for all data
import pandas as pd
lj_filelist = "/usr/src/app/radtts/data/lj_data/LJSpeech-1.1/metadata_formatted.txt"
copeland_filelist = "/usr/src/app/radtts/data/dataset-1674754128-thomas-copeland-raw/copeland/filelist.txt"
one_am_filelist = "/usr/src/app/radtts/data/dataset-1673374151-am-raw/trailer_guy_dataset/filelist.txt"
frosty_2_filelist = "/usr/src/app/radtts/data/dataset-1674764542-frosty-john-raw/dataset/filelist.txt"
general_9_filelist = "/usr/src/app/radtts/data/dataset-1674777109-general-herring-patience-raw/dataset/filelist.txt"
bertie_1_filelist = "/usr/src/app/radtts/data/dataset-1675114471-bertie-chipper-raw/dataset/filelist.txt"
general_6_filelist = "/usr/src/app/radtts/data/dataset-1675130786-general-herring-briefing-raw/dataset/filelist.txt"
frosty_5_filelist = "/usr/src/app/radtts/data/dataset-1675356555-frosty-john-cynical-raw/dataset/filelist.txt"
bertie_2_filelist = "/usr/src/app/radtts/data/dataset-1675696662-bertie-angry-raw/dataset/filelist.txt"
```

```python
# eminem_data = pd.read_csv(eminem_filelist, sep='|', index_col=None, header = None)

lj_data = pd.read_csv(lj_filelist, sep='|', index_col=None, header = None)

copeland = pd.read_csv(copeland_filelist, sep='|', index_col=None, header = None)
one_am = pd.read_csv(one_am_filelist, sep='|', index_col=None, header = None)
frosty_2 = pd.read_csv(frosty_2_filelist, sep='|', index_col=None, header = None)
general_9 = pd.read_csv(general_9_filelist, sep='|', index_col=None, header = None)
bertie_1 = pd.read_csv(bertie_1_filelist, sep='|', index_col=None, header = None)
general_6 = pd.read_csv(general_6_filelist, sep='|', index_col=None, header = None)
frosty_5 = pd.read_csv(frosty_5_filelist, sep='|', index_col=None, header = None)
bertie_2 = pd.read_csv(bertie_2_filelist, sep='|', index_col=None, header = None)
```

```python
lj_data[2] = 0
copeland[2] = 1
bertie_1[2] = 2
frosty_2[2] = 3
one_am[2] = 4
general_9[2] = 5
general_6[2] = 6
frosty_5[2] = 7
bertie_2[2] = 8
```

```python
# for i in range(eminem_data.shape[0]):
#     eminem_data.iloc[i,0] = f"eminem_14/{eminem_data.iloc[i,0]}"

for i in range(lj_data.shape[0]):
    lj_data.iloc[i,0] = f"lj_data/LJSpeech-1.1/wavs/{lj_data.iloc[i,0]}"

for i in range(copeland.shape[0]):
    copeland.iloc[i,0] = f"dataset-1674754128-thomas-copeland-raw/copeland/{copeland.iloc[i,0]}"
    
for i in range(one_am.shape[0]):
    one_am.iloc[i,0] = f"dataset-1673374151-am-raw/trailer_guy_dataset/{one_am.iloc[i,0]}"
    
for i in range(frosty_2.shape[0]):
    frosty_2.iloc[i,0] = f"dataset-1674764542-frosty-john-raw/dataset/{frosty_2.iloc[i,0]}"
    
for i in range(general_9.shape[0]):
    general_9.iloc[i,0] = f"dataset-1674777109-general-herring-patience-raw/dataset/{general_9.iloc[i,0]}"
    
for i in range(bertie_1.shape[0]):
    bertie_1.iloc[i,0] = f"dataset-1675114471-bertie-chipper-raw/dataset/{bertie_1.iloc[i,0]}"
    
for i in range(general_6.shape[0]):
    general_6.iloc[i,0] = f"dataset-1675130786-general-herring-briefing-raw/dataset/{general_6.iloc[i,0]}"
    
for i in range(frosty_5.shape[0]):
    frosty_5.iloc[i,0] = f"dataset-1675356555-frosty-john-cynical-raw/dataset/{frosty_5.iloc[i,0]}"
    
for i in range(bertie_2.shape[0]):
    bertie_2.iloc[i,0] = f"dataset-1675696662-bertie-angry-raw/dataset/{bertie_2.iloc[i,0]}"
```

```python
full_data = pd.concat([
                       lj_data, 
                       copeland,
                       bertie_1,
                       one_am,
                       frosty_2,
                       general_9,
                       general_6,
                       frosty_5,
                       bertie_2,
                      ])
```

```python
import numpy as np
MAX_WAV_VALUE=32768
from scipy.io.wavfile import write, read
from tqdm import tqdm
import librosa
output_paths =[]
for rel_path in tqdm(full_data[0]):
    rel_uuid = rel_path.split('.wav')[-2]
    output_path = f"/usr/src/app/radtts/data/{rel_uuid}_22k_normalized.wav"
    output_paths.append(output_path)
```

```python
full_data[0] = output_paths
full_data.to_csv('/usr/src/app/radtts/data/combined_022323_dirty_small_22k_normalized.txt', sep = '|', index = None, header = None)
```
